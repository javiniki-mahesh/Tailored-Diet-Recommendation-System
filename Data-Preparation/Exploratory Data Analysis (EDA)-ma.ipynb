```python
import numpy as np

# Assuming 'data' is your DataFrame
attributes = ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent',
              'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent']

height_values = []  # To store the height values

for attribute in attributes:
    heights, bins = np.histogram(data[attribute], bins=range(0, int(data[attribute].max()) + 500, 500), density=False)
    height_values.append(heights)  # Append the heights for each attribute

# 'height_values' now contains the heights for each column's histogram
for i, attribute in enumerate(attributes):
    print(f'\nAttribute: {attribute}')
    print('Bin Value\tCount')
    for bin_value, height_value in zip(bins, height_values[i]):
        print(f'{bin_value}\t\t{height_value}')
```
